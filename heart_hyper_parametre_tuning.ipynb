```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/DataSet/heart.csv')
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

HYPER PARAMETRE TUNING

```python
from sklearn.model_selection import GridSearchCV
knn1=KNeighborsClassifier()
params={'n_neighbors':[3,5,7,9],'weights':['uniform','distance'],'algorithm':['auto','ball_tree','kd_tree','brute']}
clf=GridSearchCV(knn1,params,cv=10,scoring='accuracy')
clf.fit(X_train,y_train)
```

```python
print(clf.best_params_)
```

```python
knn_new=KNeighborsClassifier(n_neighbors=3,weights='distance',algorithm='auto')
knn_new.fit(X_train,y_train)
y_pred=knn_new.predict(X_test)
y_pred
```

```python
print(classification_report(y_test,y_pred))
```
